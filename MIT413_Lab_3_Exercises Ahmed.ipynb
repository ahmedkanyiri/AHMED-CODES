*  Create a class `LinearRegression` which inherit from the `nn` (neural network) class.
*  Make sure the linear function and forward function are defined appropriately in the `LinearRegression`.
*  Again define the number of input and output dimension for your data
* Define the loss function using `MSELoss()` function
*  Define your own learning rate (`lr =  `)
* User `SDG` optimizer for your model
* Train your model by setting the `iteration_number` to 100001


```python
######### Put your code here
class LinearRegression(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(LinearRegression, self).__init__()
        
        #Linear function
        self.linear = nn.Linear(input_dim, output_dim)
        
#Forward function
    def forward(self, x):
        return self.linear(x)
```

Plot your loss function by labeling the axis appropriately
and setting an appropriate title for the graph

```python
####### Put your code here

#define the model
input_dim = 1
output_dim = 1
model = LinearRegression(input_dim, output_dim)

#define the loss function
mse = nn.MSELoss()

#set the optimization parameters
lr = 0.001
optimizer = torch.optim.SGD(model.parameters(), lr=lr)

```

```python
#Train the model
loss_list = []
iteration_num = 100001
for iteration in range(iteration_num):
    
    #optimization
    optimizer.zero_grad()
    
    #forward to get output
    results = model(yearsExperience_T)
    
    #calculate loss
    loss = mse(results, salary_T)
    
    #backward propagation
    loss.backward()
    
    #updating parameters
    optimizer.step()
    
    #store loss
    loss_list.append(loss.data)
    
    #print loss
    if (iteration % 100 == 0):
        print('epoch {}, loss{}'.format(iteration, loss.data))
```

```python
plt.plot(range(iteration_num), loss_list)
plt.xlabel('Number of Iteration')
plt.ylabel('Loss')
plt.title('Loss vs Number of Iteration')
plt.show()
```

```python
#importing necessary libraries
import os
import torch
import torchvision
from torchvision import transforms, datasets
import torch.nn as nn 
import torch.nn.functional as F
import torch.optim as optim
import matplotlib.pyplot as plt

print("Imported successfully")
```

```python
#setting device to use either gpu or cpu

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
#loading the data

data_dir = "C:\\Users\\user\\Desktop\\Chest X-ray lab 3\\xray_dataset_covid19\\"
train_dir = os.path.join(data_dir, 'train')
test_dir = os.path.join(data_dir, 'test')
```

```python
#preprocessing the data as needed

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize((0.5), (0.5))
])
```

```python
#load the dataset using torchvision

trainset = torchvision.datasets.ImageFolder(train_dir, transform=transform)
testset = torchvision.datasets.ImageFolder(test_dir, transform=transform)
```

```python
#define data loaders

train_loader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)
test_loader = torch.utils.data.DataLoader(testset, batch_size=64, shuffle=True)
```

```python
#define model

class PneumoniaClassifier(nn.Module):
    def __init__(self):
        super(PneumoniaClassifier, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3)
        self.fc1 = nn.Linear(64 * 54 * 54, 128) 
        self.fc2 = nn.Linear(128, 2)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 64 * 54 * 54)  
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
#specify loss function and optimizer

model = PneumoniaClassifier().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
#define metrics to track performance during training
epochs = 30
train_losses = []
test_losses = []
accuracy_list = []
```

```python
#training the model
for epoch in range(epochs):
    running_loss = 0
    correct = 0
    total = 0
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)
        
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        running_loss += loss.item()
    
    train_loss = running_loss / len(train_loader)
    train_losses.append(train_loss)
    
    # Evaluate on test set
    model.eval()
    test_loss = 0
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            test_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            
    test_loss /= len(test_loader)
    test_losses.append(test_loss)
    
    accuracy = 100 * correct / total
    accuracy_list.append(accuracy)
    
    print(f'Epoch [{epoch+1}/{epochs}], '
          f'Train Loss: {train_loss:.4f}, '
          f'Test Loss: {test_loss:.4f}, '
          f'Accuracy: {accuracy:.2f}%')
```

```python
#visualization the model's predictions on test images

model.eval()
with torch.no_grad():
    images, labels = next(iter(test_loader))
    images, labels = images.to(device), labels.to(device)
    outputs = model(images)
    _, predicted = torch.max(outputs, 1)
    
    fig = plt.figure(figsize=(14, 14))
    for i in range(25):
        ax = fig.add_subplot(5, 5, i+1, xticks=[], yticks=[])
        ax.imshow(images[i].cpu().permute(1, 2, 0))
        ax.set_title(f'Predicted: {predicted[i].item()}, Actual: {labels[i].item()}')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(test_losses, label='Test Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss Curve')
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(accuracy_list, label='Accuracy', color='green')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy Curve')
plt.legend()
plt.show()
```
